# Informe – Análisis de Airbnb PRAGA

## Introducción
Este proyecto analiza los alojamientos de Airbnb en Praga utilizando datos obtenidos de Inside Airbnb. El objetivo es identificar patrones en precios, valoraciones y disponibilidad, para ayudar a los viajeros a tomar decisiones informadas sobre dónde hospedarse según sus necesidades y presupuesto.

## Exploración de datos
El dataset incluye información sobre más de 10.834 alojamientos unicos con columnas como:

- Tipo de alojamiento: Alojamiento privado Casa/Apartamento, habitación privada, habitación de hotel, habitación compartida.
- Precio por noche.
- Barrio o distrito (neighbourhood).
- Valoraciones de los huéspedes (review_scores_rating).
- Disponibilidad anual (availability_365).

Durante la limpieza de datos se detectaron:

- Valores nulos en columnas de valoraciones y descripciones.
- Precios extremadamente altos que podrían ser outliers, por ejemplo apartamentos de lujo en el centro histórico.

Estas observaciones sugieren que se deben filtrar o tratar los valores atípicos antes de sacar conclusiones sobre el precio medio.

## Análisis de precios
El precio medio de los alojamientos es superior a la mediana, lo que confirma la presencia de apartamentos premium o muy caros que elevan el promedio.

Diferencias por tipo de alojamiento:
- Apartamentos completos: promedio más alto (2242 $/noche)
- Habitaciones privadas: promedio moderado (1149 $/noche)
- Habitaciones compartidas: precio más económico (605 $/noche)

Diferencias por barrio:

- Barrio Viejo (Staré Město) y Malá Strana: precios más elevados, centrados en turismo y monumentos históricos.

- Distritos periféricos (Praga 10, Praga 9): precios más bajos y mayor disponibilidad.

Recomendación para viajeros: los que busquen ubicación céntrica y cercanía a atracciones pagarán más, mientras que los distritos periféricos ofrecen ahorro y tranquilidad.

## Valoraciones
La valoración media de los alojamientos es alta. No se observa una relación fuerte
entre el precio y la valoración, lo que sugiere que un mayor precio no garantiza
mejores valoraciones.

## Disponibilidad
La disponibilidad varía entre alojamientos y barrios. Algunos tipos de alojamiento
presentan mayor disponibilidad anual, lo que podría estar relacionado con la demanda.

## Conclusiones
Explorar Airbnb en Praga es como abrir una caja de sorpresas: hay alojamientos para todos los gustos y bolsillos. Los apartamentos completos en el centro histórico pueden parecer un lujo, mientras que las habitaciones privadas y compartidas en barrios periféricos ofrecen comodidad sin romper la hucha.

Lo bueno es que, sin importar el tipo de alojamiento, las valoraciones son generalmente altas: Praga sabe cómo cuidar a sus huéspedes. La disponibilidad también varía, así que si quieres vivir la ciudad de cerca, reserva con antelación; si buscas flexibilidad y precios más tranquilos, los barrios más alejados te esperan con los brazos abiertos.

En resumen: con un poco de planificación, cualquier viajero puede encontrar su rincón perfecto en Praga, combinando comodidad, precio razonable y experiencias auténticas. Y lo mejor: siempre habrá un lugar que se ajuste a tu estilo de viaje, desde un apartamento acogedor hasta una habitación con encanto local.
